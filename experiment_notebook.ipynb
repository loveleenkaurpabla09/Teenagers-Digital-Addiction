```python
print(Vic_last_30.shape)
Vic_last_30=Vic_last_30.dropna()
print(Vic_last_30.shape)
```

```python
labels1 = Vic_last_30['Being_victim_really_affected_respondents_ability']
labels1=labels1.dropna()
labels1 = np.array(labels1)
labels1 = labels1.reshape(394,1)
labels1.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
from keras.utils import to_categorical
import numpy as np
import matplotlib.pyplot as plt


model1 = RandomForestRegressor(random_state=1, max_depth=10)
#labels1 = to_categorical(np.reshape(labels1,(428,1)))
#Vic_last_30=Vic_last_30.drop('Being_victim_really_affected_respondents_ability', axis=1)
model1.fit(Vic_last_30,labels1)

features = Vic_last_30.columns
importances = model1.feature_importances_
indices = np.argsort(importances)
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

```python
#cols1 = df.columns
finalcols = [ ]
for i in range(0,19):
    if importances[i]>=0.05:   #setting the threshold as 10%
        finalcols.append(features[i])

df1=Vic_last_30[finalcols]
df1
```

```python
from sklearn.ensemble import RandomForestRegressor
from keras.utils import to_categorical
import numpy as np
import matplotlib.pyplot as plt


model2 = RandomForestRegressor(random_state=1, max_depth=10)
#labels1 = to_categorical(np.reshape(labels1,(428,1)))
Perlast_30=Perlast_30.drop('Being_victim_really_affected_respondents_ability', axis=1)
model2.fit(Perlast_30,labels1)

features2 = Perlast_30.columns
importances2 = model2.feature_importances_
indices2 = np.argsort(importances2)
plt.title('Feature Importances')
plt.barh(range(len(indices2)), importances2[indices2], color='b', align='center')
plt.yticks(range(len(indices2)), [features2[i] for i in indices2])
plt.xlabel('Relative Importance')
plt.show()
```

```python
finalcols2 = [ ]
for i in range(0,19):
    if importances2[i]>=0.05:   #setting the threshold as 10%
        finalcols2.append(features2[i])

df2=Perlast_30[finalcols2]
df2
```

```python
from sklearn.model_selection import train_test_split

X_train1, X_test1, y_train1, y_test1 = train_test_split(df1, labels1, test_size=0.3, random_state=1)
X_train2, X_test2, y_train2, y_test2 = train_test_split(df2, labels1, test_size=0.3, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

clf1 = DecisionTreeClassifier(max_depth=4)

# Train Decision Tree Classifer
clf1 = clf1.fit(X_train1,y_train1)

#Predict the response for test dataset
y_pred1 = clf1.predict(X_test1)
print("Accuracy:",metrics.accuracy_score(y_test1, y_pred1))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

clf2 = DecisionTreeClassifier(max_depth=4)

# Train Decision Tree Classifer
clf2 = clf2.fit(X_train2,y_train2)

#Predict the response for test dataset
y_pred2 = clf2.predict(X_test2)
print("Accuracy:",metrics.accuracy_score(y_test2, y_pred2))
```
